# Gradient Calculation with AutoGrad

```python
x = torch.randn(3, requires_grad= True)
x
```

```python
y = x + 2 
y
```

This will create a node for forward prop and grad function is automatically calculated for backward prop

Gradient.PNG

```python
#Notice the back prop func
z = y * y * 2
z
```

```python
z = z.mean()
z
```

```python
z.backward() # dz/dx
x.grad
```

This calculation was successful because z was a scalar value. But if the case is different, then we will need a vector to calculate this. 

```python
x = torch.randn(3, requires_grad=True)
print(x)
y = x + 2
print(y)
z = y*y*2
print(z)
#Won't work without v
v = torch.tensor([0.1,1.0,.001], dtype=torch.float32)
z.backward(v)
print(x.grad)
```

**Gradients are saved, so have to clear it out in interations.**

```python
weights = torch.ones(4,requires_grad=True)

for epoch in range(3):
  model_output = (weights*3).sum()
  model_output.backward()
  print(weights.grad)
```

We can see that the grads are summed with the previous values. So-

```python
weights = torch.ones(4,requires_grad=True)

for epoch in range(3):
  model_output = (weights*3).sum()
  model_output.backward()
  print(weights.grad)
  weights.grad.zero_() # For clearing it out
```

**Summary**<br>
 **1) We must use requires_grad= True<br>
2) z.backward() for calculation<br>
3) weights.grad_zero_() for clearing the space**
