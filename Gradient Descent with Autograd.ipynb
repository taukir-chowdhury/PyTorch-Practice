```python
import numpy as np
import torch
```

f = w * x <br>
For example <br>
f = 2 * x

```python
X = np.array([1,2,3,4], dtype=np.float32)
Y = np.array([2,4,6,8], dtype=np.float32)
print(X,Y)
```

```python
w = 0 
```

```python
def forward(x):
  return w * x
```

```python
def loss(y, y_predicted):
  return ((y_predicted-y)**2).mean()
```

MSE = 1/N * (w*x - y)**2<br>
dj/dw = 1/N * 2x*(w*x -y) 

```python
def gradient(x,y,y_predicted):
  return np.dot(2*x,y_predicted-y).mean()
```

```python
# prediction before training
print(f'Prediction before training: f(5) = {forward(5):.3f}')
```

```python
#Training
learning_rate = 0.01
n_iter = 15
```

```python
w = 0
```

```python
for epoch in range(n_iter):
  y_pred = forward(X)
  l = loss(Y, y_pred)
  dw = gradient(X,Y,y_pred)
  w -= learning_rate * dw
  if epoch==0:
    continue
  print(f"for epoch {epoch:.3f}: w = {w:.3f} and loss = {l}")

print(f'Prediction after training: f(5) = {forward(5):.3f}')

```

# Now Everything with PyTorch

```python
X = torch.tensor([1,2,3,4], dtype=torch.float32)
Y = torch.tensor([2,4,6,8], dtype=torch.float32)
print(X,Y)
```

```python
w = torch.tensor(0.0, dtype=torch.float32, requires_grad=True)
```

```python
#Training
learning_rate = 0.01
n_iter = 40
```

```python
for epoch in range(n_iter):
  y_pred = forward(X)
  l = loss(Y, y_pred)
  l.backward() #dl/dw

  with torch.no_grad():
    w -= learning_rate * w.grad
  
  w.grad.zero_()
  if epoch % 10 == 0:
    print(f"for epoch {epoch:.3f}: w = {w:.3f} and loss = {l}")

print(f'Prediction after training: f(5) = {forward(5):.3f}')
```

# Creating the Model : replacing Forward Prop

```python
import torch
import torch.nn as nn
```

```python
#x & y have to be declared
X = torch.tensor([[1], [2], [3], [4]], dtype=torch.float32)   # 2D array where number of rows is the number of sample, column = feature
Y = torch.tensor([[2], [4], [6], [8]], dtype=torch.float32)
X
```

```python
n_samples, n_features = X.shape
print(n_samples, n_features)
```

```python
input_size = n_features
output_size = n_features
```

```python
model = nn.Linear(input_size, output_size)
```

```python
#because model needs a tensor
X_test = torch.tensor([5], dtype=torch.float32)
```

```python
#item() is used for getting the exact float value
print(f'Prediction before training f(5) = {model(X_test).item():.3f}')
```

```python
learning_rate = 0.5
n_iters = 200
```

```python
#Optimizer is updated
loss = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr = learning_rate)
```

```python
w = torch.tensor([0], dtype=torch.float32, requires_grad=True)
w
```

```python
for epoch in range(n_iters):
  #forward pass
  y_pred = model(X)

  #loss
  l = loss(Y, y_pred)

  #gradients dl/dw
  l.backward()

  #update
  optimizer.step()

  #zero_gradient
  optimizer.zero_grad()


  if epoch % 10 == 0:
    [w,b] = model.parameters()
    print(f'epoch {epoch + 1}: w = {w[0][0].item():.3f} loss = {l:.3f}')

print(f'Prediction before training f(5) = {model(X_test).item():.3f}')


```

# Writing the model manually

```python
X = torch.tensor([[1], [2], [3], [4]], dtype=torch.float32)   # 2D array where number of rows is the number of sample, column = feature
Y = torch.tensor([[2], [4], [6], [8]], dtype=torch.float32)
X_test = torch.tensor([5], dtype=torch.float32)
n_samples, n_features = X.shape
```

```python
w = torch.tensor([0], dtype=torch.float32, requires_grad=True)
```

```python
learning_rate = 0.01
n_iters = 100
```

```python
input_size = n_features
output_size = n_features
```

```python


class LinearRegression(nn.Module):
  def __init__(self,input_dim, output_dim):
    super(LinearRegression, self).__init__()

    #define layers
    self.lin = nn.Linear(input_dim, output_dim)

  def forward(self,x):
    return self.lin(x)
```

```python
w = torch.tensor([0], dtype=torch.float32, requires_grad=True)
w
```

```python
model = LinearRegression(input_size,output_size)
```

```python
#item() is used for getting the exact float value
print(f'Prediction before training f(5) = {model(X_test).item():.3f}')
```

```python
#Optimizer is updated
loss = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr = learning_rate)
```

```python
for epoch in range(n_iters):
  #forward pass
  y_pred = model(X)

  #loss
  l = loss(Y, y_pred)

  #gradients dl/dw
  l.backward()

  #update
  optimizer.step()

  #zero_gradient
  optimizer.zero_grad()


  if epoch % 10 == 0:
    [w,b] = model.parameters()
    print(f'epoch {epoch + 1}: w = {w[0][0].item():.3f} loss = {l:.3f}')

print(f'Prediction after training f(5) = {model(X_test).item():.3f}')


```
